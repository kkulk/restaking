```python
import requests
import pandas as pd

operator_stats_url = "https://u--1.s3.us-east-2.amazonaws.com/eigenlayer-operator-stats.json"
operator_registrations_url = "https://u--1.s3.us-east-2.amazonaws.com/eigenlayer-operator-registrations.json"
avs_stats_url = "https://u--1.s3.us-east-2.amazonaws.com/eigenlayer-avs-stats.json"
native_tvl_url = "https://u--1.s3.us-east-2.amazonaws.com/eigenlayer-native-tvl.json"

def fetch_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        response.raise_for_status()

operator_stats = fetch_data(operator_stats_url)
operator_registrations = fetch_data(operator_registrations_url)
avs_stats = fetch_data(avs_stats_url)
native_tvl = fetch_data(native_tvl_url)

print("Native TVL Data:")
print(native_tvl)

operator_stats_df = pd.DataFrame(operator_stats['value'])
operator_registrations_df = pd.DataFrame(operator_registrations['value'])
avs_stats_df = pd.DataFrame(avs_stats['value'])

if isinstance(native_tvl, list):
    native_tvl_df = pd.DataFrame(native_tvl)
else:
    native_tvl_df = pd.DataFrame([native_tvl])

print("Operator Stats DataFrame:")
print(operator_stats_df.head())

print("\nOperator Registrations DataFrame:")
print(operator_registrations_df.head())

print("\nAVS Stats DataFrame:")
print(avs_stats_df.head())

print("\nNative TVL DataFrame:")
print(native_tvl_df.head())

operator_stats_df.to_csv('operator_stats.csv', index=False)
operator_registrations_df.to_csv('operator_registrations.csv', index=False)
avs_stats_df.to_csv('avs_stats.csv', index=False)
native_tvl_df.to_csv('native_tvl.csv', index=False)
```
