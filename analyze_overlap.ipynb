```python
AVS_dist = {}
for _, row in avs.iterrows():
    metadata = eval(row['metadata'])
    share = eval(row['strategyAndShare'])
    AVS_dist[metadata['name']] = dict([(k, int(v)) for k, v in share.items()])
```

```python
AVS_dist
```

```python
overlap = {}
for name1, dist1 in AVS_dist.items():
    for name2, dist2 in AVS_dist.items():
        if name1 == name2: continue
        k1, k2 = set(dist1.keys()), set(dist2.keys())
        k = k1.intersection(k2)
        d1_overlap = np.sum([dist1[a] for a in k])
        d2_overlap = np.sum([dist2[a] for a in k])
        d1_total = np.sum([dist1[a] for a in k1])
        d2_total = np.sum([dist2[a] for a in k2])
        overlap[(name1, name2)] = (d1_overlap/float(d1_total), d2_overlap/float(d2_total))
```

```python
overlap
```

```python
raw_data = np.concatenate([v for _, v in overlap.items()])
```

```python
cnts, bins = np.histogram(raw_data)
```

```python
plt.plot(0.5*(bins[1:]+bins[:-1]), cnts)
```

```python
mean_overlap = {}
for name in AVS_dist.keys():
    total = 0.0
    cnt = 0
    for (n1, n2), (d1, d2) in overlap.items():
        if name == n1:
            total += d1
            cnt += 1
    mean_overlap[name] = float(total)/float(cnt)
```

```python
mean_overlap
```

```python
std_overlap = {}
for name in AVS_dist.keys():
    ol = []
    for (n1, n2), (d1, d2) in overlap.items():
        if name == n1:
            ol.append(d1)
    
    std_overlap[name] = np.array(ol).std()
```
